### The Signum Function
$$
\mathrm{sgn}(t) = \begin{Bmatrix}
 1,& t>0\\ 
 0,& t=0\\ 
-1,& t<0
\end{Bmatrix}
$$

```python
def sgn(t):
    x = np.sign(t)
    return x
```

```python
t = np.linspace(-10, 10, 1000)
x = sgn(t)
```

```python
plt.figure(figsize=(8,3))
plt.plot(t, x, label='sgn')
plt.xlabel('$\mathrm{time}(t)$', fontsize=12)
plt.ylabel('$\mathrm{sgn}(t)$', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### The Unit-Step Function
$$
\mathrm{u}(t) = \begin{Bmatrix}
 1,& t>0\\ 
 1/2,& t=0\\ 
0,& t<0
\end{Bmatrix}
$$

```python
def us(t):
    '''
    define function unit step
    '''
    y = np.heaviside(t, 0.5)
    return y
```

```python
t = np.linspace(-10, 10, 1000)
x = us(t)
```

```python
plt.figure(figsize=(8,3))
plt.plot(t, x, label='unit step')
plt.xlabel('$\mathrm{time}(t)$', fontsize=12)
plt.ylabel('$\mathrm{u}(t)$', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### The Unit-Ramp Function
$$
\mathrm{ramp}(t)
\begin{Bmatrix}
 t& t>0\\ 
 0& t<=0
\end{Bmatrix}
=\int_{-\infty}^{t} \mathrm{u}(\lambda)=t \mathrm{u}(t)
$$

```python
def ramp(t):
    '''
    define function ramp
    '''
    x = t * us(t)
    return x
```

```python
t = np.linspace(-10, 10, 1000)
x = ramp(t)
```

```python
plt.figure(figsize=(8,3))
plt.plot(t, x, label='unit ramp')
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$\mathrm{ramp}(t)$', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### The Unit-Area Rectangular Pulse of width $\Delta$ 

```python
def rect(t):
    '''
    define function rect
    '''
    delta = 1
    x = 1/delta * (us(t+delta/2) - us(t-delta/2))
    return x
```

```python
t = np.linspace(-2, 2, 1000)
x = rect(t)
```

```python
plt.figure(figsize=(8,3))
plt.plot(t, x, label='unit rect')
plt.xlabel('$\mathrm{time}(t)$', fontsize=12)
plt.ylabel('$\mathrm{rect}(t)$', fontsize=12)
plt.legend(loc='upper right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### The Unit-Impulse Function
#### The unit-impulse function is taken from the unit-area rectangle pulse as $\Delta$ approaches zero

```python
def impulse(t):
    '''
    define function rect
    '''
    delta = 0.01 # small number
    x = 1/delta * (us(t+delta/2) - us(t-delta/2))
    return x
```

```python
t = np.linspace(-2, 2, 1000) # changable according to delta
x = impulse(t)
```

```python
plt.figure(figsize=(8,3))
plt.plot(t, x, label='unit rect')
plt.xlabel('$\mathrm{time}(t)$', fontsize=12)
plt.ylabel('$\mathrm{rect}(t)$', fontsize=12)
plt.legend(loc='upper right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

#### The Sampling Property of the Impulse Function
$$
\int_{ -\infty }^{ \infty } \mathrm{g}(t) \mathrm { \delta } (t-t_0) dt = \mathrm{g}(t_0)
$$

#### The Scaling Property of the Impulse Function
$$
\delta(a(t-t_0))=\frac{1}{\left | a \right |} \delta(t-t_0)
$$

#### The Unit Periodic Impulse or Impulse Train
$$
\mathrm { \delta }_T = \sum_{n = -\infty }^{ \infty } \mathrm { \delta } (t - nT)
$$

### Example Graphing function combinations
$$ \mathrm x_1(t) = e^{-t} \sin (20 \pi t) + e^{-t/2} \sin (19 \pi t) $$
$$ \mathrm x_2(t) = \mathrm{rect}(t) \cos (20 \pi t) $$

```python
t = np.arange(0, 6, 1/240)
x1 = np.exp(-t) * np.sin(20 * math.pi * t) + np.exp(-t/2) * np.sin(19 * math.pi * t)
```

```python
plt.figure(figsize=(8,3))
plt.plot(t, x1, label='unit rect')
plt.xlabel('$\mathrm{time}(t)$', fontsize=12)
plt.ylabel('$ \mathrm{x_1}(t)$', fontsize=12)
#plt.legend(loc='upper right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

```python
t = np.arange(-2, 2, 1/240)
x2 = rect(t) * np.cos(20 * math.pi * t)
```
