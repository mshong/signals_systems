```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import scienceplots
import math
```

```python
plt.style.use(['science', 'notebook', 'grid'])
```

- descrete unit impulse function

```python
def impD(n):
    y = np.heaviside(n, 1) - np.heaviside(n-1, 1)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = impD(n)
```

```python
plt.figure(figsize=(10,3))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$\delta (n)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-10,10))
plt.xticks([i+1 for i in x_position])
plt.show()
```

- discrete unit step function

```python
def usD(n):
    y = np.heaviside(n, 1)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = usD(n)
```

```python
plt.figure(figsize=(10,3))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$u(n)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-10,10))
plt.xticks([i+1 for i in x_position])
plt.show()
```

- discrete unit sign function

```python
def signD(n):
    y = np.sign(n)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = signD(n)
```

```python
plt.figure(figsize=(10,5))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$sign(n)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-10,10))
plt.xticks([i+1 for i in x_position])
plt.show()
```

- discrete unit ramp function

```python
def rampD(n):
    y = n * usD(n)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = rampD(n)
```

```python
plt.figure(figsize=(10,5))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$ramp(n)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-10,10))
plt.xticks([i+1 for i in x_position])
plt.show()
```
