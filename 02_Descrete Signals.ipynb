```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import scienceplots
import math
```

```python
plt.style.use(['science', 'notebook', 'grid'])
```

```python
def impD(n):
    y = y = np.heaviside(n, 1) - np.heaviside(n-1, 1)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = impD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$u(t)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

```python
def usD(n):
    y = np.heaviside(n, 1)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = usD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$u(t)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

```python
def signD(n):
    y = np.sign(n)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = signD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$u(t)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

```python
def rampD(n):
    y = n * usD(n)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = rampD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$u(t)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```
