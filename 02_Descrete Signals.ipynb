### The Unit-Impulse Function
$$
\ \delta[n] = \left\{\begin{matrix}
 1,& n=0\\ 
 0,& n \neq 0
\end{matrix}\right.
$$

```python
def impD(n):
    y = y = np.heaviside(n, 1) - np.heaviside(n-1, 1)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = impD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$ \mathrm{\delta} [n]$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### The Unit-Sequence Function
$$
\ \mathrm{u}[n] = 
\left\{\begin{matrix}
 1,& n \geq 0\\ 
 0,& n \neq 0
\end{matrix}\right.
$$

```python
def usD(n):
    y = np.heaviside(n, 1)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = usD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$ \mathrm{u} [n]$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### The Signum Function
$$
\ \mathrm{sgn}[n] = 
\left\{\begin{matrix}
 1,& n \gt 0\\
 0,& n = 0 \\
 -1,& n \lt 0
\end{matrix}\right.
$$

```python
def signD(n):
    y = np.sign(n)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = signD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$ \mathrm{u} [n]$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### The Unit-Ramp Function
$$
\ \mathrm{ramp}[n] = 
\begin{Bmatrix}
 n,& n \geq 0\\ 
 0,& n \lt 0
\end{Bmatrix}
=n \mathrm{u}(t)
$$

```python
def rampD(n):
    y = n * usD(n)
    return y
```

```python
n = np.arange(-10, 10, 1)
x = rampD(n)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$u(n)$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

### Example Graphing and scaling of descrete-tim functions
$$
\mathrm g [n] = 10(0.8)^n  \sin (3 \pi n/16) \mathrm u[n]
$$

```python
def g(n):
    #y = 10.*(0.8)**n * usD(n)
    y = 10.*(0.8)**n * np.sin(3*np.pi*n/16.)* usD(n)
    return y
```

```python
n = np.arange(-5, 48, 1)
g0 = g(n)
g1 = g(2*n)
g2 = g(n/3)
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, g0)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$ \mathrm{g}[n]$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.stem(n, g1)
plt.xlabel('$n$', fontsize=12)
plt.ylabel('$ \mathrm{g}[2n]$', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```
