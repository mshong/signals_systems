- unit step function

```python
def us(x):
    '''
    define function unit step
    '''
    y = np.heaviside(x, 0.5)
    return y
```

```python
t = np.linspace(-5, 5, 1000)
x = us(t)
```

```python
plt.figure(figsize=(10,3))
plt.plot(t, x, label='unit step')
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$u(t)$', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-5,5))
plt.xticks([i+1 for i in x_position])
plt.show()
```

- unit ramp function

```python
def ramp(x):
    '''
    define function ramp
    '''
    y = x * us(x)
    return y
```

```python
t = np.linspace(-5, 5, 1000)
x = ramp(t)
```

```python
plt.figure(figsize=(10,3))
plt.plot(t, x, label='unit ramp')
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$ramp(t)$', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-5,5))
plt.xticks([i+1 for i in x_position])
plt.show()
```

- unit rectangle function

```python
def rect(x):
    '''
    define function rect
    '''
    y = us(x+0.5) - us(x-0.5)
    return y
```

```python
t = np.linspace(-2, 2, 1000)
x = rect(t)
```

```python
plt.figure(figsize=(10,3))
plt.plot(t, x, label='unit rect')
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$rect(t)$', fontsize=12)
plt.legend(loc='lower right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-2,2))
plt.xticks([i+1 for i in x_position])
plt.show()
```

### Example 2.1
$$ \mathrm x_1(t) = e^{-t} \sin (20 \pi t) + e^{-t/2} \sin (19 \pi t) $$
$$ \mathrm x_2(t) = \mathrm{rect}(t) \cos (20 \pi t) $$

```python
t = np.arange(0, 6, 1/240)
x1 = np.exp(-t) * np.sin(20 * math.pi * t) + np.exp(-t/2) * np.sin(19 * math.pi * t)
```

```python
plt.figure(figsize=(10,5))
plt.plot(t, x1, label='unit rect')
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$ x_1(t)$', fontsize=12)
#plt.legend(loc='upper right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(0,5))
plt.xticks([i+1 for i in x_position])
plt.show()
```

```python
t = np.arange(-2, 2, 1/240)
x2 = rect(t) * np.cos(20 * math.pi * t)
```

```python
plt.figure(figsize=(8,3))
plt.plot(t, x2, label='unit rect')
plt.xlabel('$t$', fontsize=12)
plt.ylabel('$x_2(t)$', fontsize=12)
#plt.legend(loc='upper right', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
x_position = list(range(-2,2))
plt.xticks([i+1 for i in x_position])
plt.show()
```
